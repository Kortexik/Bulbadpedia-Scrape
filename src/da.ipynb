```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('../saves/ModelDataset.csv')
df2 = pd.read_csv('../saves/PokemonsForModel.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
cr = df2['Catch Rate']
```

```python
df = pd.concat([df, cr], axis=1)
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(), xticklabels=True, yticklabels=True)
```

```python
sns.scatterplot(x='Catch Rate', y='Total Base Stat', data=df)
```
